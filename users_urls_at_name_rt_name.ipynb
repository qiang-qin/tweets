```python
from collections import defaultdict
import csv
import re
import sys
def flushPrint(s):
    sys.stdout.write('\r')
    sys.stdout.write('%s' % s)
    sys.stdout.flush()
    
from collections import defaultdict
import csv
 
data_dict3 = defaultdict(list) # error
error_num = 0
line_num = 0
total_num = 0
 
bigfile = open('D:/Data/ows/ows-raw.txt', 'rb')
chunkSize = 10000000
chunk = bigfile.readlines(chunkSize)
while chunk:
    total_num += len(chunk)
    lines = csv.reader((line.replace('\x00','') for line in chunk), delimiter=',', quotechar='"')
    for i in lines:
        line_num+=1
        if line_num%1000000==0:
            flushPrint(line_num)
        try:
            rt_patterns = re.compile(r"(RT|via)((?:\b\W*@\w+)+)", re.IGNORECASE)
            rt_names = rt_patterns.findall(i[1])
            date = i[3]
            for rt_name in rt_names:
                if len(date)==10:
                    data_dict3[date].append(rt_name)
        except Exception, e:
            print e
            error_num+=1
            pass
    chunk = bigfile.readlines(chunkSize) 
print line_num, total_num,error_num
```
